```python
import os
import matplotlib.pyplot as plt
```

```python
# Define the directory path
directory = r"E:\IITP\Academics\7th Semester 26 Jul'23 - 1 Dec'23\BTP\PHM Society Bearings Data\Data\IMS\2nd_test\2nd_test"

# Initialize an empty list to store the averages
averages = []
```

```python
# Iterate through each file in the directory
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    
    # Read the file and calculate the average of the first column
    with open(filepath, 'r') as file:
        column_sum = 0
        num_rows = 0
        for line in file:
            # Assuming columns are tab-separated
            values = line.split('\t')
            if len(values) > 0:
                column_sum += float(values[0])  # Summing up the values in the first column
                num_rows += 1  # Counting the number of rows
    
        # Calculate the average of the first column for the current file
        if num_rows > 0:
            file_average = column_sum / num_rows
            averages.append(file_average)
```

```python
# Create a list of sample indices from 600 to the end
index = list(range(len(averages)))

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Plot the data
plt.plot(index[1:], averages[1:])

plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Average Magnitude of Vibration of each file for Test 2 Bearing 1')
plt.show()
```

```python
k = 800

# Create a list of sample indices from 600 to the end
index = list(range(k, len(averages)))

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Plot the data
plt.plot(index, averages[k:])

plt.xlabel('File Index')
plt.ylabel('Average')
plt.title('Average of Each Column')
plt.show()
```

```python
# Define the file path for saving the predicted probabilities
output_file = "magnitude_averages.txt"

# Write the predicted probabilities to the text file
with open(output_file, 'w') as f:
    for prob in averages:
        f.write(str(prob) + '\n')

print("Saved to", output_file)
```
