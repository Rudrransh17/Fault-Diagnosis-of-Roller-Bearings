```python
import os
import matplotlib.pyplot as plt
```

```python
# Define the directory path
directory = r"E:\IITP\Academics\7th Semester 26 Jul'23 - 1 Dec'23\BTP\PHM Society Bearings Data\Data\IMS\1st_test\1st_test"

# Initialize an empty list to store the averages
averages_b3 = []
averages_b4 = []
```

```python
# Iterate through each file in the directory
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    
    # Read the file and calculate the average of the first column
    with open(filepath, 'r') as file:
        b3_sum = 0
        b3_rows = 0
        b4_sum = 0
        b4_rows = 0
        for line in file:
            # Assuming columns are tab-separated
            values = line.split('\t')
            if len(values) > 0:
                b3_sum += float(values[4])  # Summing up the values in the third column
                b3_rows += 1  # Counting the number of rows
                b4_sum += float(values[6])  # Summing up the values in the third column
                b4_rows += 1  # Counting the number of rows
    
        # Calculate the average of the first column for the current file
        if b3_rows > 0:
            b3_file_avg = b3_sum / b3_rows
            averages_b3.append(b3_file_avg)
        
        if b4_rows > 0:
            b4_file_avg = b4_sum / b4_rows
            averages_b4.append(b4_file_avg)
```

```python
# Create a list of sample indices from 600 to the end
index = list(range(len(averages_b3)))

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Plot the data
plt.plot(index[1:], averages_b3[1:])

plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Average Magnitude of Vibration of each file for Test 1 Bearing 3')
plt.show()
```

```python
# Create a list of sample indices from 600 to the end
index = list(range(len(averages_b4)))

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Plot the data
plt.plot(index[1:], averages_b4[1:])

plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.title('Average Magnitude of Vibration of each file for Test 1 Bearing 4')
plt.show()
```
